```python
import pandas as pd
import numpy as np
import scipy.io
import matplotlib.pyplot as plt

data = scipy.io.loadmat('CO2data.mat')
data
ytrain = data['ytrain']
ytest = data['ytest']
xtest = data['xtest']
xtrain = data['xtrain']
print(ytrain.shape)
print(ytest.shape)
print(xtrain.shape)
print(xtest.shape)
```

```python
from datetime import datetime
t = [datetime(2016,1,1), datetime(2017,12,31)]
fig, ax = plt.subplots()
ax.plot(xtrain, ytrain, label='Train')
ax.plot(xtest, ytest, label='Test')
ax.legend()
ax.set_title('Co2 data')
ax.set_xlabel('Year')
ax.set_ylabel('Co2 concentration, ppm')
fig.show()
```
